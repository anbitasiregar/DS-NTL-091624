## Task 14: Create a new dictionary

Create a dictionary called `song_dictionary` which consists of each track name `string` as a key and a `tuple` of artists associated with each track as a value.

```python
# call_on_students(1)
```

```python
# Your code here:

song_dictionary = {}

for song in first_twenty_songs:
    song_name = song["name"]
    artists = [artist["name"] for artist in song["artists"]]

    song_dictionary[song_name] = tuple(artists)

song_dictionary

# alternative
# {song["name"]: tuple(artist["name"] for artist in song["artists"]) for song in songs}
```

## Task 15: Write a function

Create a function with takes an **artist name** and the **song_dictionary** as arguments, and returns a `list` of songs written by that artist. 

```python
# call_on_students(1)
```

```python
# Easier to do things outside of a function first
# Let's try for The Weeknd

def songs_by_artist(artist_name, song_dictionary):
    song_names = []

    # loop through song_dictionary
    for song, artist_tuple in song_dictionary.items():
        if artist_name in artist_tuple:
            song_names += [song]

    # return list of songs written by that artist
    return song_names

print(songs_by_artist("The Weeknd", song_dictionary))
```

```python
# Your code here

def find_song_by_artist(artist_name, song_dict):
    
    '''
    Parameters:
    arist_name: a string of an artist's name to be used to search the dictionary
    song_dict:  a dictionary of top_twenty songs with song name as keys and a list of 
    artist names as values
    
    Returns:
    A list of songs which the given artist appeared on
    '''
    
    song_names = []

    # loop through song_dictionary
    for song, artist_tuple in song_dict.items():
        if artist_name in artist_tuple:
            song_names += [song]

    # return list of songs written by that artist
    return song_names

# alternative
# [song for song in song_dictionary if artist_name in song_dict[song]]
```
